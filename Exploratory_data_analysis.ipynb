Finding the missing values inside the features

```python
def features_with_missing_values(df):
    for i in df.columns:
        if df[i].isna().sum()>0:
            print('The Feature ',i,' has '+ str(df[i].isna().sum()) + ' missing values')
            
features_with_missing_values(df)
```

Now we will check the corelation between our dataset using seaborn library

```python
corrmap=df.corr()
top=corrmap.index
graph=sns.heatmap(df[top].corr())
```

```python
#The graph coming here is small and color is not very helping so we will change that
corrmap=df.corr()
plt.figure(figsize=(25,25))
top=corrmap.index
graph=sns.heatmap(df[top].corr(),cmap="Blues")
```
